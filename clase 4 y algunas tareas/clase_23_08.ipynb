```python
def pisar_elemento(M,e):
    M_shape = M.shape
    lista_M = M.reshape(M.size)
    copia_lista = []
    for i in lista_M:
        if i == e:
            copia_lista.append(-1)
        else:
            copia_lista.append(i)
    salida_lista = np.array(copia_lista)
    salida = salida_lista.reshape(M_shape)
    return salida

ejemplo_M = np.array([[[0,1,2,3],[4,5,6,7],[2,2,2,2]],[[2,2,1,2],[3,3,2,4,],[5,6,7,2]]])
e = 2

print(ejemplo_M)
print(pisar_elemento(ejemplo_M, e))
```

```python
import pandas as pd
import os

archivo = 'arbolado-en-espacios-verdes.csv'
fname = os.path.join(archivo)
data_arboles_parques = pd.read_csv(fname)
data_jacaranda = data_arboles_parques[data_arboles_parques["nombre_com"] == "Jacarandá"]
data_palos_borrachos = data_arboles_parques[data_arboles_parques["nombre_com"] == "Palo borracho"]

altura_jacaranda = data_jacaranda["altura_tot"]
diametro_jacaranda = data_jacaranda["diametro"]

print("datos Jacaranda")
print(f"cantidad de arboles = {data_jacaranda.count()[0]}")
print(f"altura maxima = {altura_jacaranda.max()} mts ")
print(f"altura minima = {altura_jacaranda.min()} mts")
print(f"altura promedio = {altura_jacaranda.mean()} mts")
print(f"diametro maxima = {diametro_jacaranda.max()} mts ")
print(f"diametro minima = {diametro_jacaranda.min()} mts")
print(f"diametro promedio = {diametro_jacaranda.mean()} mts")

altura_palo_borracho = data_palos_borrachos["altura_tot"]
diametro_palo_borracho = data_palos_borrachos["diametro"]

print("datos palos borrachos")
print(f"cantidad de arboles = {data_palos_borrachos.count()[0]}")
print(f"altura maxima = {altura_palo_borracho.max()} mts ")
print(f"altura minima = {altura_palo_borracho.min()} mts")
print(f"altura promedio = {altura_palo_borracho.mean()} mts")
print(f"diametro maxima = {diametro_palo_borracho.max()} mts ")
print(f"diametro minima = {diametro_palo_borracho.min()} mts")
print(f"diametro promedio = {diametro_palo_borracho.mean()} mts")

data_arboles_parques
```
